```python
#1. Create inheritance using MobilePhone as base class and Apple & Samsung as child class
class MobilePhone:
    def __init__(self, dual_sim, front_camera, rear_camera, ram, storage):
        self.screen_type = "Touch Screen"
        self.network_type = "4G/5G"
        self.dual_sim = dual_sim
        self.front_camera = front_camera
        self.rear_camera = rear_camera
        self.ram = ram
        self.storage = storage

    def phone_specs(self):
        print(f"Screen Type   : {self.screen_type}")
        print(f"Network Type  : {self.network_type}")
        print(f"Dual SIM      : {self.dual_sim}")
        print(f"Front Camera  : {self.front_camera}")
        print(f"Rear Camera   : {self.rear_camera}")
        print(f"RAM           : {self.ram}")
        print(f"Storage       : {self.storage}")
class Apple(MobilePhone):
    def __init__(self):
        # Apple typically doesn't support dual SIM (in older models)
        super().__init__(dual_sim=False, front_camera="12MP", rear_camera="48MP", ram="4GB", storage="64GB")
class Samsung(MobilePhone):
    def __init__(self):
        # Samsung often supports dual SIM
        super().__init__(dual_sim=True, front_camera="16MP", rear_camera="32MP", ram="3GB", storage="32GB")
print("Apple Phone Specs:")
apple_phone = Apple()
apple_phone.phone_specs()

print("\nSamsung Phone Specs:")
samsung_phone = Samsung()
samsung_phone.phone_specs()
```

```python
#2. Create basic mobile phone functionalities in the classes like: make_call, recieve_call, take_a_picture, etc.
class MobilePhone:
    def __init__(self, dual_sim, front_camera, rear_camera, ram, storage):
        self.screen_type = "Touch Screen"
        self.network_type = "4G/5G"
        self.dual_sim = dual_sim
        self.front_camera = front_camera
        self.rear_camera = rear_camera
        self.ram = ram
        self.storage = storage

    def phone_specs(self):
        print(f"Screen Type   : {self.screen_type}")
        print(f"Network Type  : {self.network_type}")
        print(f"Dual SIM      : {self.dual_sim}")
        print(f"Front Camera  : {self.front_camera}")
        print(f"Rear Camera   : {self.rear_camera}")
        print(f"RAM           : {self.ram}")
        print(f"Storage       : {self.storage}")

    def make_call(self, number):
        print(f"Dialing {number}...")

    def receive_call(self, from_contact):
        print(f"Incoming call from {from_contact}")

    def take_a_picture(self, camera_type="rear"):
        if camera_type == "front":
            print(f"Click! Photo taken with {self.front_camera} front camera.")
        elif camera_type == "rear":
            print(f"Click! Photo taken with {self.rear_camera} rear camera.")
        else:
            print("Invalid camera type. Choose 'front' or 'rear'.")

class Apple(MobilePhone):
    def __init__(self):
        super().__init__(dual_sim=False, front_camera="12MP", rear_camera="48MP", ram="4GB", storage="64GB")
class Samsung(MobilePhone):
    def __init__(self):
        super().__init__(dual_sim=True, front_camera="16MP", rear_camera="32MP", ram="3GB", storage="32GB")

print("=== Apple Phone ===")
iphone = Apple()
iphone.phone_specs()
iphone.make_call("9876543210")
iphone.receive_call("Mom")
iphone.take_a_picture("front")

print("\n=== Samsung Phone ===")
galaxy = Samsung()
galaxy.phone_specs()
galaxy.make_call("8123456789")
galaxy.receive_call("Friend")
galaxy.take_a_picture("rear")
```

```python
#3. Use super() constructor for calling parent classâ€™s constructor
class MobilePhone:
    def __init__(self, brand, dual_sim, front_camera, rear_camera, ram, storage):
        self.brand = brand
        self.screen_type = "Touch Screen"
        self.network_type = "4G/5G"
        self.dual_sim = dual_sim
        self.front_camera = front_camera
        self.rear_camera = rear_camera
        self.ram = ram
        self.storage = storage

    def phone_specs(self):
        print(f"Brand         : {self.brand}")
        print(f"Screen Type   : {self.screen_type}")
        print(f"Network Type  : {self.network_type}")
        print(f"Dual SIM      : {self.dual_sim}")
        print(f"Front Camera  : {self.front_camera}")
        print(f"Rear Camera   : {self.rear_camera}")
        print(f"RAM           : {self.ram}")
        print(f"Storage       : {self.storage}")

class Apple(MobilePhone):
    def __init__(self):
        # Using super() to call the base class constructor
        super().__init__(
            brand="Apple",
            dual_sim=False,
            front_camera="12MP",
            rear_camera="48MP",
            ram="4GB",
            storage="128GB"
        )

class Samsung(MobilePhone):
    def __init__(self):
        # Using super() to call the base class constructor
        super().__init__(
            brand="Samsung",
            dual_sim=True,
            front_camera="16MP",
            rear_camera="32MP",
            ram="6GB",
            storage="64GB"
        )
iphone = Apple()
galaxy = Samsung()
print("=== Apple iPhone Specs ===")
iphone.phone_specs()
print("\n=== Samsung Galaxy Specs ===")
galaxy.phone_specs()
```
